```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
b = 12.0
mu_d = 8.0
c_d = 1.0
rho = 0.05

beta = 0.95
R_kinri = 1 / beta

hosei = 0.1

Q = np.array([[0.0, 0.0, 0.0], [0.0, hosei, 0.0], [0.0, 0.0, 0.0]])
A = np.array([[1.0, 0.0, 0.0], [-b, R_kinri, 1.0], [(1.0 - rho) * mu_d, 0.0, rho]])
B = np.array([[0.0], [1.0], [0.0]])
C = np.array([[0.0], [0.0], [c_d]])
R_hyouka = np.array([[1]])


iter_t = 100
# iter_arr = [-40]
iter_arr = [-1e-5, -9]

d_0 = 0
y = np.zeros((1, 3))

# IID gaussian distribution
eps = np.random.randn(iter_t)

for sig in iter_arr:
    # print(y)
    y = np.insert(y, 0, np.array([1, 0, d_0]), axis=0)
    F, K, P, Pt = olrprobust(beta, A, B, C, Q, R_hyouka, sig)

    print(f"F={F}")
    print(f"K={K}")
    print(f"P={P}")
    print(f"Pt={Pt}")

    Tmp = A - np.dot(B, F)
    # print(Tmp)
    for i in range(iter_t):
        # print(i)
        # F, K, P, Pt = olrprobust(beta, A, B, C, Q, R_hyouka, sig)
        # Tmp = A -np.dot(B,F) + np.dot(C,K)
        # print(Tmp)
        y = np.insert(y, 0, np.dot(Tmp, y[0].T) + C.T * eps[i], axis=0)

k_get = y[:, 1]
d_get = y[:, 2]
k_get = k_get[::-1]
d_get = d_get[::-1]
one_get = y[:, 0]
x = np.array(range(iter_t + 1))
```

```python
plt.figure()
plt.xlabel("t")
plt.ylabel("k")
plt.title("k")
for where, sig in enumerate(iter_arr):
    #%print("where=%d"%where)
    plt.plot(
        x,
        k_get[(iter_t + 1) * where + 1 : (iter_t + 1) * where + 1 + iter_t + 1],
        label="sig=%f" % sig,
    )
    plt.legend(prop={"size": 10})
plt.savefig("..\k_kuri_in2")
```

```python
plt.figure()
plt.xlabel("t")
plt.ylabel("d")
plt.title("d")
for where, sig in enumerate(iter_arr):
   
    plt.plot(
        x,
        d_get[(iter_t + 1) * where + 1 : (iter_t + 1) * where + 1 + iter_t + 1],
        label="sig=%d" % sig,
    )
    plt.legend(prop={"size": 10})
plt.savefig("..\d_kuri_in2")
```

```python
plt.figure()
plt.xlabel("t")
plt.ylabel("c")
plt.title("c")
for where, sig in enumerate(iter_arr):
    c = np.zeros(iter_t+1)
    # print("c size before %d"%c.shape)
    c[0] = 0
    for i in range(1, iter_t+1):
        #print("k = %d"%k_get[(iter_t + 1) * where + i - 1])
        c[i] = (
            R_kinri * k_get[(iter_t + 1) * where + i -1]
            + d_get[(iter_t + 1) * where + i]
            - k_get[(iter_t + 1) * where + i]
        )
        #print("c = %d"%c[i])
        if c[i] <-800:
            print("whe=%d,sig=%d,i=%d"%(where,sig,i))
    plt.plot(
        x,
        c,
        label="sig=%d" % sig,
    )
    plt.legend(prop={"size": 10})
plt.savefig("..\c_kuri_in2")
```
